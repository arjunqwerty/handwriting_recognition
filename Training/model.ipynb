```python
train_img = glob("feature_extraction/1_enhanced_images/*.png")[1000:1009]
fig, axes = plt.subplots(3, 3, figsize=(6, 4))
for i, image in enumerate(train_img):
    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i//3][i%3].imshow(img)

axes[0][1].set_title('Train Images',fontsize=17)
plt.show()
```
